```python
income_data = df1[['Country Code', 'IncomeGroup']]
region_count = income_data['IncomeGroup'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(region_count, labels=region_count.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Percentage of Countries in Income Categories')
plt.show()
```
